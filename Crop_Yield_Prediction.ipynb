```python
#Question-2
Area_mid=data["Area"].median()
Production_mid = data["Production"].median()

data["Area"]=data["Area"].fillna(Area_mid).astype(int)
data["Production"]=data["Production"].fillna(Production_mid)
```

```python
#Question-3
data["Annual_Rainfall"]=data["Annual_Rainfall"].fillna(data["Annual_Rainfall"].mean()).astype(int)
```

```python
#Question-4
Avg_val=(data["Fertilizer"].max()+ data["Fertilizer"].min())/2

data["Fertilizer"]=data["Fertilizer"].fillna(Avg_val)
```

```python
#Question-5
Q1=data["Yield"].quantile(0.25)
Q3=data["Yield"].quantile(0.75)

data["Yield"]=data["Yield"].fillna(data["Yield"].median())
```

```python
#Question 6 & 7 Both combaind
data["Production_per_Area"]=data["Production_per_Area"].ffill()
```

```python
data.isnull().sum()
```

```python
Crop_Production = [
    "Record_ID","Crop_Year","Season","State","Area","Production",
    "Yield","Production_per_Area","Yield_Category","Crop"
 ]
data[Crop_Production].to_csv("Crop_Production.csv", index=False)


Climate_Agriculture = [
   "Record_ID","Annual_Rainfall","Rainfall_Category","Fertilizer",
     "Fertilizer_per_Area","Pesticide","Pesticide_per_Area","Year_Group"
 ]
data[Climate_Agriculture].to_csv('Climate_Agriculture.csv', index=False)
```
